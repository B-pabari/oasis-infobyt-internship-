```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Create new features if needed (e.g., month, year, day of week)
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Day_of_Week'] = df['Date'].dt.dayofweek
```

```python
df.describe()
```

```python
# Plot sales over time
plt.figure(figsize=(3,6))
plt.plot(df['Price per Unit'], df['Quantity'])
plt.xlabel('Price per Unit')
plt.ylabel('Quantity')
plt.title('Sales Over Time')
plt.show()

# Seasonal Decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(df['Quantity'], model='additive', period=12)  # Adjust period as needed
result.plot()
plt.show()
```

```python
# Customer Analysis
customer_sales = df.groupby('Date')['Transaction ID'].sum().sort_values(ascending=False)
customer_sales.head(10).plot(kind='bar')
plt.xlabel('Date')
plt.ylabel('Transactinal id')
plt.title('Top 10 Customers by Sales')
plt.show()
```

```python
# Product Analysis
product_sales = df.groupby('Quantity')['Price per Unit'].sum().sort_values(ascending=False)
product_sales.head(10).plot(kind='bar')
plt.xlabel('Quantity')
plt.ylabel('Total Price per Unit')
plt.title('Top 10 Products by Sales')
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
